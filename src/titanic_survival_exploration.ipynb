## Getting Started
In this lab, you will see how decision trees work by implementing a decision tree in sklearn.

We'll start by loading the dataset and displaying some of its rows.

```python
# Import libraries necessary for this project
import numpy as np
import pandas as pd
from IPython.display import display # Allows the use of display() for DataFrames

# Pretty display for notebooks
%matplotlib inline

# Set a random seed
import random
random.seed(42)

# Load the dataset
# goto your drive and just copy path where your dataset is availabe 
in_file = '/content/gdrive/MyDrive/colabspace/MachineLearningProject/src/datasets/titanic_data.csv'
full_data = pd.read_csv(in_file)
# Print the first few entries of the RMS Titanic data
display(full_data.head())
```

Recall that these are the various features present for each passenger on the ship:
- **Survived**: Outcome of survival (0 = No; 1 = Yes)
- **Pclass**: Socio-economic class (1 = Upper class; 2 = Middle class; 3 = Lower class)
- **Name**: Name of passenger
- **Sex**: Sex of the passenger
- **Age**: Age of the passenger (Some entries contain `NaN`)
- **SibSp**: Number of siblings and spouses of the passenger aboard
- **Parch**: Number of parents and children of the passenger aboard
- **Ticket**: Ticket number of the passenger
- **Fare**: Fare paid by the passenger
- **Cabin** Cabin number of the passenger (Some entries contain `NaN`)
- **Embarked**: Port of embarkation of the passenger (C = Cherbourg; Q = Queenstown; S = Southampton)

Since we're interested in the outcome of survival for each passenger or crew member, we can remove the **Survived** feature from this dataset and store it as its own separate variable `outcomes`. We will use these outcomes as our prediction targets.  
Run the code cell below to remove **Survived** as a feature of the dataset and store it in `outcomes`.

```python
# Store the 'Survived' feature in a new variable and remove it from the dataset
outcomes = full_data['Survived']
features_raw = full_data.drop('Survived', axis = 1)

# Show the new dataset with 'Survived' removed
display(features_raw.head())
```

The very same sample of the RMS Titanic data now shows the **Survived** feature removed from the DataFrame. Note that `data` (the passenger data) and `outcomes` (the outcomes of survival) are now *paired*. That means for any passenger `data.loc[i]`, they have the survival outcome `outcomes[i]`.

## Preprocessing the data

Now, let's do some data preprocessing. First, we'll one-hot encode the features.

```python
features = pd.get_dummies(features_raw)
```

And now we'll fill in any blanks with zeroes.

```python
features = features.fillna(0.0)
display(features.head())
```

## Training the model

Now we're ready to train a model in sklearn. First, let's split the data into training and testing sets. Then we'll train the model on the training set.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, outcomes, test_size=0.2, random_state=42)
```

```python
# Import the classifier from sklearn
from sklearn.tree import DecisionTreeClassifier

# Define the classifier, and fit it to the data
model= DecisionTreeClassifier()
model.fit(X_train,y_train)
```

## Testing the model
Now, let's see how our model does, let's calculate the accuracy over both the training and the testing set.

```python
# Making predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calculate the accuracy
from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print('The training accuracy is', train_accuracy)
print('The test accuracy is', test_accuracy)
```

## Confusion Matrix

[Refrence](https://wiki.pathmind.com/accuracy-precision-recall-f1#:~:text=That%20is%2C%20a%20good%20F1,total%20failure%20when%20it's%200%20.)

1.  **Accuracy** : accuracy tell us that wether model is trained correctly with x% percent.

2. **Precision** : Precision says, what propocertion of the positive identification are actually correct?

3. **Recall (sensitivity)** : Recall says, what propocertion of actual positive are identified correctly? it's also called **senitivity**, true positive rate and hit rate.
4. **F1 score** : F1 score measure the overall accuracy that combines precition and recall. Good F1 score mean, low **false positive rate** and low false negative rate.
Note:- recall measure the **true postive rate**. F1 score **1** means perfect model and F1 score **0** mean failure model.

5. **Specificity** :  Specificity also known as **true negative rate**, say that what proportion of actual negatives are correctly identify.

## Predict `y values` using cross validation and evalute model using confusion matrix

```python
from sklearn.model_selection import cross_val_predict

y_train_pred = cross_val_predict(model, X_train, y_train, cv = 5)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_train, y_train_pred)
```

In above, confusion_matrix return a confusion matrix in which each rows represent **actual value** and each columns represent **predicted values**.

see below:- 
                                          
                                          Predicted values
                                         
                                         No              Yes
                          --------------------------------------
                              No         TN               FP
              Actual values

                             Yes        FN               TP



### Calculate Accuracy, precision, Recall and Specifisity

```python
tn, fp, fn, tp = confusion_matrix(y_train, y_train_pred).ravel()

tn, fp, fn, tp
```

```python
# Wow, Accuracy, precision are too good. accuracy tell us the accuracy of correct classify of true postive and true negative.

accuracy = (tp + tn) / (tp + tn + fp + fn)
accuracy
```

```python
precision = tp / (tp + fp)
precision
```

```python
# Recall is also known as sentitivity , `true positive rate`

recall = tp / (tp + fn)
recall
```

```python
# F1 score is also good with 71% that tell us our model has perfectly train for generalization

F1_score = 2 * (precision * recall) / (precision + recall)
F1_score
```

```python
# Specificity is also known as `true negative rate`
specificity = tn / (tn + fp)
specificity
```

In model evaluation of the classification problem, the main goal is, the has good true positive rate with keeping low false positive rate. you can see our model is as expected.

```python
# we can calculate false positive rate 1 minus true negative rate
false_posive_rate = 1 - specificity
false_posive_rate
```
